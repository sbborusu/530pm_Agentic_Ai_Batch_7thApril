# Uniform Distribution
*   Every value in a given range has an equal chance of occurring.
## Example: Rolling a Fair Die
*   A standard die has 6 faces: {1, 2, 3, 4, 5, 6}.
*   When you roll it, each number has an equal chance of appearing.
*   If you plot this, you'll see a flat bar chart because all outcomes are equally likely.
*   Falls under univariate analysis where one variable is analyzed at a time.


## np.random.rand(100)
*   Generates 100 random floating point numbers between 0 and 1, drawn from a uniform distribution.
*   Every number between 0 and 1 has an equal probability of being chosen.
*   It does not accept parameters to change the range directly. If you want a different range, lets say between 0 and 100, multiply the rand(100) array results with 100.














```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(10000)

# Observe the histogram at the top and it roughly matches the flat line
plt.hist(x, bins=10, edgecolor="white")
plt.axhline(y=1000, color="r", linestyle="-")
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.show()
```

# Normal Distribution

1. Mean (μ): The center of the distribution (where the peak occurs).
2. Standard deviation (σ): Controls the spread (how wide or narrow the bell curve is).
3. Symmetry: The curve is symmetric around the mean.
4. 68-95-99.7 Rule:
About 68% of values fall within μ ± 1σ.
About 95% within μ ± 2σ.
About 99.7% within μ ± 3σ.
5. Shape: Bell-shaped curve, tails extend infinitely but probabilities become very small.

## Example

Let's say we want a normal distribution with:

1. Mean (μ) = 50
2. Standard deviation (σ) = 10

This could represent something like test scores where:

1. Most students score near 50.
2. Scores between 40 and 60 are common (±1σ).
3. Scores below 20 or above 80 are rare (±3σ).

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 50
sigma = 10
n = 10000

# Generate samples
np.random.seed(42)
x = np.random.normal(mu,sigma,n)
# Below line of code can also be used via standard normal distribution approach which is explained in the next section.
#x = mu + sigma * np.random.randn(n)

# Plot histogram with theoretical PDF
plt.figure(figsize=(8, 4), dpi=150)
plt.hist(x, bins=40, density=True, color="#4c78a8", alpha=0.7, edgecolor="white")

# Overlay theoretical normal PDF
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x_vals - mu)/sigma)**2)
plt.plot(x_vals, pdf, color="red", lw=2, label=f"Normal PDF (μ={mu}, σ={sigma})")

# Labels
plt.title("Normal Distribution Example (μ=50, σ=10)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# Insights from the plot

# A bell curve centered at 50.
# Most values between 40 and 60.
# Very few values below 20 or above 80.
```

# Standard Normal Distribution
*   It's a bell-shaped probability distribution that is symmetric around its mean.
*   Most values cluster near the center (mean), and fewer values occur as you move away from the mean.
*   np.random.randn() function below explains more details about characteristics of 'standard normal distribution'
## np.random.randn(100)
*   For randn(), the default normal distribution has:
1.   Mean (μ) = 0
2.   Standard deviation (σ) = 1
3.   Most values cluster near 0.
4.   About 68% of values fall within ±1.
5.   About 95% of values fall within ±2.
6.   Extreme values like ±3 are rare (only about 0.3% of samples).
7.   the 68–95–99.7 rule







```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 samples from standard normal distribution
np.random.seed(42)
x = np.random.randn(10000)

# Plot histogram with density=True
plt.hist(x, bins=40, density=True, color="#4c78a8", alpha=0.7, edgecolor="white")

# Overlay theoretical normal PDF
x_vals = np.linspace(-4, 4, 500)
pdf = 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x_vals**2)
plt.plot(x_vals, pdf, color="red", lw=2, label="Theoretical Normal PDF")
plt.title("Standard Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
