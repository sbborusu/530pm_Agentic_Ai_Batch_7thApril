```python
import matplotlib.pyplot as plt
import numpy as np

# Define x-values
x1 = np.array([0, 1, 2, 3, 4])
x2 = np.array([1, 2, 3, 4, 5])  # shifted x-values for scenario 3

# Scenario 1: Same slope, different intercepts
y1_1 = 2 * x1 + 1
y1_2 = 2 * x1 + 4

# Scenario 2: Same intercept, different slopes
y2_1 = 1 * x1 + 2
y2_2 = 3 * x1 + 2

# Scenario 3: Same slope and intercept, different x-values
y3_1 = 2 * x1 + 1
y3_2 = 2 * x2 + 1

# Scenario 4: Same intercept and same starting x-value
y4_1 = 2 * x1 + 3
y4_2 = 2 * x1 + 3

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot Scenario 1
axs[0, 0].plot(x1, y1_1, label='y = 2x + 1')
axs[0, 0].plot(x1, y1_2, label='y = 2x + 4')
axs[0, 0].set_title('Same Slope, Different Intercepts')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot Scenario 2
axs[0, 1].plot(x1, y2_1, label='y = 1x + 2')
axs[0, 1].plot(x1, y2_2, label='y = 3x + 2')
axs[0, 1].set_title('Same Intercept, Different Slopes')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot Scenario 3
axs[1, 0].plot(x1, y3_1, label='y = 2x + 1 (x1)')
axs[1, 0].plot(x2, y3_2, label='y = 2x + 1 (x2)')
axs[1, 0].set_title('Same Slope & Intercept, Different x-values')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot Scenario 4
axs[1, 1].plot(x1, y4_1, label='y = 2x + 3')
axs[1, 1].plot(x1, y4_2, '--', label='y = 2x + 3 (same line)')
axs[1, 1].set_title('Same Slope, Intercept, and x-values')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the slope
m = 1

# Define two different intercepts
c1 = 1
c2 = 1

# Generate x values
x = np.linspace(1, 5, 10)

# Calculate y values for both lines
y1 = m * x + c1
y2 = m * x + c2

# Plot the lines
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = 2x + 1', color='blue', marker='o')
plt.plot(x, y2, label='y = 2x + 4', color='green', marker='o')

# Add labels and title
plt.title('Effect of Changing Intercept (c) in y = mx + c')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.array([0, 1, 2, 3, 4])

# Define y values for two lines with different slopes but same starting point
# y = mx + c
# c is '2' and same for both, so the intercept or starting point for both lines is 2.
y1 = 1 * x + 2
y2 = 3 * x + 2

print('x: ', x)
print('y1: ', y1)
print('y2: ', y2)
# Plotting the lines
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Slope = 1', marker='o')
plt.plot(x, y2, label='Slope = 3', marker='o')
plt.title('Effect of Changing Slope with Same Starting Point')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Define x values
x = [0, 1, 2, 3, 4]

# Define y values for two lines with different slopes but same starting point
# Starting point is (0, 2)
y1 = [2 + 1 * xi for xi in x]  # slope = 1
y2 = [2 + 3 * xi for xi in x]  # slope = 3

print('x: ', x)
print('y1: ', y1)
print('y2: ', y2)

# Plotting the lines
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Slope = 1', marker='o')
plt.plot(x, y2, label='Slope = 3', marker='o')
plt.title('Effect of Changing Slope with Same Starting Point')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```
