```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Function to update the plot
def update(num, x, y, line):
    line.set_ydata(np.sin(x + num / 10.0))
    return line,

# Create an animation
ani = animation.FuncAnimation(fig, update, frames=100, fargs=[x, y, line], interval=50)
plt.title('Animated Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Synthetic data: Monthly sales over a year
months = np.arange(1, 13)
sales = np.array([120, 150, 170, 160, 180, 200, 220, 210, 230, 250, 270, 260])

# Create a trend line (linear fit)
trend_line = np.poly1d(np.polyfit(months, sales, 1))
print(trend_line)

# Extrapolate future values (next 6 months)
future_months = np.arange(13, 19)
future_sales = trend_line(future_months)
print(future_sales)

# Plotting the data points with markers
plt.plot(months, sales, marker='o', label='Monthly Sales')

# Plotting the trend line
plt.plot(months, trend_line(months), linestyle='--', color='red', label='Trend Line')

# Plotting the extrapolated future values
plt.plot(future_months, future_sales, linestyle=':', marker = '>', color='green', label='Extrapolated Sales' )

# Adding gridlines
plt.grid(True)

# Adding title and labels
plt.title('Monthly Sales Over a Year with Trend Line and Extrapolation')
plt.xlabel('Month')
plt.ylabel('Sales')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
