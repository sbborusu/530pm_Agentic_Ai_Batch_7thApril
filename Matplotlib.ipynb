```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Function to update the plot
def update(num, x, y, line):
    line.set_ydata(np.sin(x + num / 10.0))
    return line,

# Create an animation
ani = animation.FuncAnimation(fig, update, frames=100, fargs=[x, y, line], interval=50)
plt.title('Animated Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
days = ['M', 'T', 'W', 'Th', 'F', 'S', 'Su']
months = [30,20,40,50,30,10,20]
plt.plot(days, months, 'o--b')

# Draw horizontal lines from y-axis tickers touching the markers
for i in range(len(days)):
  plt.axhline(y=months[i], color='black', linestyle='-')
plt.title('Time Spent Reading')
plt.xlabel('Days')
plt.ylabel('Months')
plt.show()
```

# Line Plotting

```python
# Line plot without markers
import matplotlib.pyplot as plt
import numpy as np

# Plotting line from (1,3) to (4,9) to (8, 10) using arrays
xpoints = np.array([1, 4, 8])
ypoints = np.array([3, 9, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Line plot with markers and without line
import matplotlib.pyplot as plt
import numpy as np

# Plotting line from (1,3) to (4,9) to (8, 10) using arrays
xpoints = np.array([1, 4, 8])
ypoints = np.array([3, 9, 10])

# 'o' indicates the marker, but since nothing was mentioned as second parameter ie., linestyle was not mentioned,
# it will be treated as empty resulting in no line at all
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```
