```python
import sympy
```

```python
J, w = sympy.symbols('J w')
```

```python
#J = w ** 2
#J = w ** 3
#J = w
#J = 1/w
J = (1/2) * (w ** 2)
print (J)
```

```python
dJ_dw = sympy.diff(J, w)
print (dJ_dw)
```

```python
dJ_dw.subs([[w,2]])
```
