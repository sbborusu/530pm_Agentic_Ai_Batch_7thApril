```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import numpy as np
```

# Generate a synthetic non-linear dataset using a sine function


```python
np.random.seed(42)
X = np.sort(5 * np.random.rand(500, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0]) # Add noise
```

# Linear Regression Model


```python
linear_model = LinearRegression()
linear_model = linear_model.fit(X,y)
y_linear_pred = linear_model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_linear_pred)
plt.show()
```

# Polynomial regression model

```python
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly,y)

y_poly_predict = poly_model.predict(X_poly)
```

# Plotting

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='gray', alpha=0.5, label='Original Data')
plt.plot(X, y_linear_pred, label='Linear Regression')
plt.plot(X, y_poly_predict,  label='Polynomial Regression (Degree 3)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear vs Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

# Model Evaluation



```python

r2_linear = r2_score(y, y_linear_pred)
rmse_linear = mean_squared_error(y, y_linear_pred)

r2_poly = r2_score(y, y_poly_predict)
rmse_poly = mean_squared_error(y, y_poly_predict)

# The Polynomial Regression model (degree 3) captures the non-linear sine pattern much more effectively than the Linear Regression model.
# The significantly higher R and lower RMSE confirm that the data has a non-linear shape.
# This validates the approach of using model-based diagnostics to determine the nature of the data.
# Print the results
print("Linear Regression:")
print(f"R2 Score: {r2_linear:.4f}")
print(f"RMSE: {rmse_linear:.4f}\n")

print("Polynomial Regression (Degree 3):")
print(f"R2 Score: {r2_poly:.4f}") # Close to '1' and better value compared to linear regression
print(f"RMSE: {rmse_poly:.4f}") # Less errors compared to linear regression
```
